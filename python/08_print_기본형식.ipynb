# print()

## 1. 기본 문법
- print(value, ..., sep='구분기호', end='\n', flush=False)

```python
a = 5
a
```

```python
b = 3
b
```

```python
a + b
```

```python
print(a, end='\t')
print(b, end='\t')
print(a+b)
```

```python
print (1, 2, 3, sep=',', flush=True) # 1-2-3
```

```python
name = '홍길동'
age = 30
```

```python
print('이름: ', name, '나이: ', age)
```

```python
# 접합연산자(+)를 이용하여 문자열 변환
print('이름:'+name, '나이:'+str(age)) # int()-숫자로 변환, str()-문자열로 변환
```

```python
# % 포맷(old version)
print("이름: %s, 나이: %d" % (name, age))
```

```python
# format{} 포맷(new version)
print("이름: {}, 나이: {}".format(name, age))
```

```python
# f-string
print(f"이름: {name}, 나이: {age}")
```

```python
# 여러 값을 출력 콤마(,) 구분
print(1, 2, 3)
print(1, 2, 3, sep = ',')
```

```python
# 줄바꿈 없이 옆으로 출력
lst = [1, 2, 3, 4, 5]
for i in lst :    # 0, 1, 2, 3, 4
    print(i, end= '\t')
```

```python
#빈 줄 출력
print()
```

```python
# 8을 50번 출력
print('*' * 50)
```
