# file 열기(open())

- 내장 함수인 open() 함수를 이용하여 파일을 연 다음, Read, Write이 가능함
- 형식 : file = open(파일주소)
- 속성
  - r : 읽기(read) 모드로 파일 열기(기본)
  - w : 쓰기(write) 모드로 파일 열기
  - a : 추가(append) 모드로 파일 열기, 같은 이름의 파일 없으면 w 모드로 생성됨
  - b : 바이너리 파일 모드로 파일 열기
  - t : 텍스트 파일 모드로 파일 열기

# 1. 파일 쓰기

- 파일 쓰기를 하려면 open() 함수의 쓰기 모드로 열어야 함
- 파일 쓰기가 종료하면, 반드시 파일을 닫아야 함

```python
# myFile.txt 파일을 생성
f = open('myFile.txt', 'w')
f.write('Hello~ myFile~ ^^')
f.close()
```

# 2. 파일 읽기

- 파일 쓰기를 하려면 open(r) 함수의 쓰기 모드로 열어야 함
- 파일 쓰기가 종료하면, 반드시 파일을 닫아야 함

```python
# myFile.txt 파일 읽기
f = open('myFile.txt', 'r')
f_text = f.read()
f.close()

f_text
```

```python
# myFile_ko.txt 파일 생성(인코딩)
f = open('myFile_ko.txt', 'w', encoding='cp949')
f.write('안녕~ 파이썬 ^^')
f.close()

# myFile_ko.txt 파일 읽기
f = open('myFile_ko.txt', 'r')
# print(f.read())
text = f.read()
f.close()

text
```

## 3. 파일에 문자열 한 줄씩 쓰기

```python
# 파일에 문자열 한 줄씩 쓱
f = open('myFile_multi.txt', 'w')
for num in range(1, 6):
    str = f"num: {num} \n"
    f.write(str)
f.close()
```

```python
# myFile_multi.txt 읽기
f = open('myFile_multi.txt', 'r')
line1 = f.readline()
print(line1)
# print(f.read())
f.close()
```

```python
# myFile_multi.txt 파일을 반복적으로 읽기
f = open('myFile_multi.txt')
line = f.readline()
while line :
    print(line, end = "")
    line = f.readline()
print("-"*10)
f.close()
```

- readlines() : 파일 전체를 읽어 리스트 타입으로 반환

```python
f = open('myFile_multi.txt', 'r')
lines = f.readlines()
print(lines)
```

```python
# lines 객체를 for 문으로 출력
for line in lines:
    print(line)
```
