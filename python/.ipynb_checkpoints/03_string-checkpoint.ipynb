# 문자열

- 문자열 : 파이썬에서는 따옴표('', "") 로 둘러싼 문자 집합을 의미

## 1. 문자열 생성

- 시작과 끝에 같은 따옴표를 사용한다. "~~~", '~~~'

```python
print('Hello~ Python')
```

```python
print("Hello~ Python")
```

```python
str1 = "Hello~ Python"
str2 = 'Hello~ Python'
print(str1, str2)
```

```python
print(type(str1), type(str2))
```

```python
# 제 이름은 '홍길동' 입니다.
str3 = "제 이름은 '홍길동' 입니다."
str3
```

```python
# 문자열을 여러 줄 입력하는 방법
str4 = """파이썬을
배우는 
중입니다"""
print(str4)
```

# 2. 문자열 다루기

- 더하기 연산자(+)와 곱하기 연산자(*)
  - (+) 연산자 : 문자열 결합(concatenation)
  - (*) 연산자 : 곱합 수 만큼 문자열 반복됨

```python
a = "Hello~"
b = "Python!!"
c = a + b
print(c)
c
```

```python
a * 3 + b * 3
```

```python
(a + b) * 3
```

```python
c * 3
```

```python
print(a)
print(b)
print('-'*30)
print(c)
```
