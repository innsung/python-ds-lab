# 2. Dataframe 생성

### (1) 리스트(list)로 생성

```python
## 파이썬의 2차원 리스트 생성생성
list1 = [['김철수', 28, '개발자'], ['이영희', 34, '디자이너']]

# 리스트를 데이터프레임으로 생성
list_df = pd.DataFrame(list1, columns=['이름', '나이','직업'])
list_df
```

### (2) 딕셔너리(dict)로 생성

```python
# 파이썬의 딕셔너리를 이용하여 데이터프레임 생성
dic1 = {'Name': ['이순신', '유관순','김유신','홍길동'],
        'Year': [2011, 2016, 2015, 2015],
        'Gender': ['Male', 'Female', 'Male', 'Male']
       }
dic1
```

```python
# 딕셔너리로 데이터프레임 생성
dict_df = pd.DataFrame(dic1)
dict_df
```

```python
# dict_df 데이터프레임에 컬럼명 추가, 인덱스 변경
dict_df2 = pd.DataFrame(dic1, 
                        columns=['Name','Year','Gender','Age','Address'],
                        index=['a','b','c','d'])
dict_df2
```

### (3) ndarray로 생성

```python
array1 = np.array([['김철수', 30, '개발자'], ['이영희', 20, '디자이너']])
array1
```

```python
array_df = pd.DataFrame(array1, columns=['Name', 'Age', 'Job'], index=['one','two'])
array_df
```

# 6. value_counts()
- 동일한 개별 데이터 값이 몇건이 있는지 정보를 제공

```python
tt_value_counts = titanic_df['Survived'].value_counts()
print(tt_value_counts, type(tt_value_counts))
```

```python
# 두 개 이상의 컬럼을 출력하는 경우, 2차원 형태로 입력
tt_value_counts = titanic_df[['Survived', 'Pclass']].value_counts()
print(tt_value_counts, type(tt_value_counts))
```
