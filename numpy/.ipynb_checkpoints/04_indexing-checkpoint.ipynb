# 인덱싱(indexing)

- ndarray에서 데이터를 검색하는 방법
- 인덱싱 방법
|종류 | 방법 | 특징 |
|-----|------|------|
|단일값 인덱싱|arr[2]|특정위치 단일값 출력|
|슬라이싱|arr[1:4]|연속된 범위|
|팬시인덱싱|arr[[0, 2, 4]]|불연속 위치 다중값 출력|
|불린인덱싱|arr[arr>5]|조건에 맞는 값 출력|

### (1) 단일값 인덱싱

```python
import numpy as np
```

```python
#1~9까지 ndarray 데이터 생성
arr = np.arange(1, 10)
arr
```

```python
# 2번지의 데이터 출력
print(type(arr))
print(type(arr[2]))

print(arr.dtype)
print(arr[2].dtype)
print(arr.shape)
```

```python
# 인덱스별 단일값 출력
print(arr)
print(arr[0])
print(arr[2])
print(arr[5])
print(arr[-1])
print(arr[-2])
```

### (2) 값 변경

```python
arr[0]  = 100.0
print(arr)
arr[-1] = 900
print(arr)
```

### (3) 2차원 배열
- 2차원 배열은 [행, 열] 기준으로 검색

```python
arr = np.arange(1, 10)
arr
```

```python
# 3행 3열의 2차원 배열로 변환
arr2 = np.reshape(arr, (3, 3))
arr2
```

```python
# 3 출력
print(arr2.shape)
print(arr2[0, 2])
# 8출력
print(arr2[2, 1])
# 5출력
print(arr2[1, 1])
#마지막 행, 마지막 열 출력
print(arr2[-1, -1])
```

#### (4) 슬라이싱
- ':' 기호를 이용하여 연속된 데이터를 출력

```python
array1 = np.arange(1, 10)
array1
```

```python
# 0, 1, 2 세 게의 데이터 출력
array1[:3]
```

```python
# 마지막 데이터 3개 출력
array1[-1:-4:-1]
```

```python
array1
#[4, 5, 6]
print(array1[3:6])
#[1, 2, 3]
print(array1[:3])
#[1, 3, 5, 7 ,9]
print(array1[::2])
#[9, 8 ,7...1]역순으로 출력
print(array1[::-1])
# 마지막 데이터 3개 출력
print(array1[-3:])
```

- 2차원 슬라이싱

```python
# 3행 3열의 2차원 배열 생성, 데이터 1~9
arr = np.arange(1, 10).reshape((3, 3))   #reshape(object, shape...)
arr
```

```python
print(arr[0:1, 0:3]) #[[1 2 3]]
print(arr[0:2, 0:2])
print(arr[1:3, 1:3])
print(arr[:2, :2])
print(arr[2:, :3])
```

```python
print(arr[0], arr[0].shape) # 첫번째 로우 출력, 1차원 배열
print(arr[2], arr[2].shape) # 첫번째 로우 출력, 1차원 배열
print(arr[:2]) # 첫번째 로우 출력, 1차원 배열
```

### (5) 팬시 인덱싱
- 불연속적인 여러 위치의 값을 인덱싱하여 출력

- 1차원 배열의 팬시 인덱싱

```python
array1 = np.arange(1, 10)
array1
```

```python
print(array1[0])   # 단일값 인덱싱
print(array1[0:3]) # 슬라이싱
print(array1[[0, 2, 3]])  # 팬시 인덱싱
```

```python
# 슬라이싱 + 팬시 인덱싱
value = array1[4:][[2, 4]]
print(value, value.shape)
```

- 2차원 배열 팬시 인덱싱

```python
# (3, 3) 1~9
arr = np.arange(1, 10).reshape((3, 3))
arr
```

```python
# [[행], [열]]
print(arr[[0, 1], 2])
print(arr[[0, 2], 2])
print(arr[[0, 2], [1]])
print(arr[[0, 2], [0, 2]]) #(0,0), (2,2)
print(arr[[0, 1, 2], 1]) #(0,1), (1,1), (2,1)
```

### (6) 불린 인덱싱
- 조건식을 사용하여 인덱싱
- 데이터 필터링에 자주 사용됨

```python
arr = np.arange(1, 10)
arr
```

```python
# 1차원 배열 arr의 데이터 중 5보다 큰 데이터만 출력
arr[arr>5]
```

```python
# 1차원 배열 arr의 데이터 중 5보다 작은 데이터만 출력
arr[arr<5]
```

```python
# 불린 인덱싱은 내부적으로 조건에 맞는 불린값으로 필터링을 실행한 후
# True 값만 출력
arr_bool = arr > 5
print(arr)
print(arr_bool, arr[arr_bool], arr[arr>5])
```

```python
# arr 배열의 짝수 값 출력
print(arr[arr % 2 == 0])
print(arr[arr % 2 == 1])
```

```python
# 2의 배수 또는 5의 배수 출력, 단 |, & 연산 실행 시 조건을 괄호()로 묶어야 함
print(arr[(arr % 2 == 0) | (arr % 5 == 0)])
```

- 불린인덱싱 vs for 루프

```python
arr = np.arange(1, 10)
arr
```

```python
# 불린 인덱싱(효율적, 권장)
arr1 = arr[arr > 5]
print(arr1, type(arr1))
```

```python
# for 루프 => [6, 7, 8, 9]
target = []
for n in range(1, len(arr)+1):
    if n > 5:
        target.append(n)
print(target, type(target))
arr2 = np.array(target)
print(arr2, type(arr2))
```

```python
# 학생 성적 데이터에서 80점 이상 학생 이름 추출
names = np.array(['홍길동','김유신','강감찬','이순신','스미스'])
scores = np.array([72, 81, 95, 89, 88])
# 80점 이상 학생 이름 출력
print(names[scores>=80])
# 80점 이상인 점수 출력
print(scores[scores>=80])
```
